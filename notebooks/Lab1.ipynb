Подготовим все для перевода номинальных долларов в реальные.

```python
from bs4 import BeautifulSoup
import requests
```

```python
url = "http://www.usinflationcalculator.com/inflation/consumer-price-index-and-annual-percent-changes-from-1913-to-2008/"
r = requests.get(url)
data = r.text
soup = BeautifulSoup(data, 'html.parser')
table = soup.find('table')
rows = table.tbody.findAll('tr')
years = []
cpis = []

for row in rows:
    year = row.findAll('td')[0].get_text()
    if year.isdigit() and int(year) < 2023:
        years.append(int(year))
        cpis.append(float(row.findAll('td')[13].get_text()))

cpi_table = pd.DataFrame({
    "year" : years,
    "avg_annual_cpi" : cpis
})

cpi_table.head(10)
```

Определим функцию для перевода номинальных долларов в реальные доллары в 2022 году (2023 год не закончился и значения avg_annual_cpi не имеет)

```python
def get_real_value(nominal_amt, old_cpi, new_cpi):
    real_value = (nominal_amt * new_cpi)/ old_cpi
    return real_value
    
```

Установим константу для наших реальных долларов

```python
CPI_2022 = float( cpi_table [ cpi_table [ 'year' ] == 2022][ 'avg_annual_cpi' ])
```

Преобразуем значения столбцов budget и gross для каждого фильма в реальные долларовые эквиваленты 2007г:

```python
real_domestic_gross = []
real_budget_values = []

for index, row in df.iterrows():
    gross = row['gross']
    budget = row['budget']
    year = row['title_year']
    cpi = float(cpi_table[cpi_table['year'] == int(year)]['avg_annual_cpi'])
    if gross == 'unknown':
        real_gross = 0
    else:
        real_gross = get_real_value(gross, cpi, CPI_2022)
    if budget == 'unknown':
        real_budget = 0
    else:
        real_budget = get_real_value(budget, cpi, CPI_2022)
    real_domestic_gross.append(real_gross)
    real_budget_values.append(real_budget)

df["real_domestic_gross"] = real_domestic_gross
df["real_budget"] = real_budget_values
```

Теперь мы готовы к проверкам гипотез

Список гипотез:
1. Бюджет, потраченный за год на фильмы, больше с каждым годом.
2. В зависимости от системы рейтинга контента и значения рейтинга, заметно разнятся количества лайков фильмов в facebook
3. Фильмы из USA в прокате получают больше денег чем из UK
4. Более популярны фильмы, включающие жанр Action

Проверим первую гипотезу, для этого сгрупируем фильмы по годам и сложим бюджеты фильмов каждого года по отдельности.

```python
df_year = df.groupby('title_year')['real_budget'].sum()
print(df_year)
```

Судя по известным данным из таблицы гипотеза ложная, но если посмотреть по количеству фильмов за каждый год и количеству фильмов, дата которых не известна, то можно сказать что эта оценка не совсем объективна.

```python
df_year_count = df.groupby('title_year')['real_budget'].count()
print(df_year_count)
```

Проверим вторую гипотезу, для этого сгруппируем фильмы по возрастному рейтингу и сложим лайки

```python
df.loc[df['movie_facebook_likes'] == 'unknown', 'movie_facebook_likes'] = 0
new = []
for i in df['movie_facebook_likes']:
    if i !=0:
        i = int(i[:-1])
        new.append(i)
    else:
        new.append(i)
df['movie_facebook_likes'] = new
df_PG_likes = df.groupby('content_rating')['movie_facebook_likes'].sum()
print(df_PG_likes)
```

Как видно, в зависимости от рейтинга самого фильма и системы рейтиногов сильно отличаются лайки в facebook

Проверим последнюю гипотезу, для этого найдем фильмы, в жанрах которых есть Action, и найдем средний imdb рейтинг
Создав копию таблицы заменим жанры фильмов, где есть Action на True, а где нет на False

```python
df_action= df
df_action['genres'] = df_action['genres'].apply(lambda x: True if 'Action' in x.split('|') else False)
```

Посмотрим какие значения теперь находятся в жанрах

```python
sorted(df_action.genres.unique())
```

Как видим все значения жанров обработались, теперь сгруппируем их и найдем средние imbi-score

```python
df_action_res = df_action.groupby('genres')['imdb_score'].mean()
print("Action :",df_action_res[True])
print("No Action :",df_action_res[False])
```

Как видно гипотеза ложная. Фильмы без жанра Action более популярны по среднему значению imdb-score
