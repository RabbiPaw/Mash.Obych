Выбросы будут влиять на правдоподобность данных и читаемость гистограм

```python
data.head(10)
```

```python
data.describe()[['year_of_release','audience_count','imdb_rating','tomatometer_rating','audience_rating','tomatometer_count']]
```

```python
columns = ['year_of_release','audience_count','imdb_rating','tomatometer_rating','audience_rating','tomatometer_count']

for column in columns:
    data.hist(column)
```

По гистограммам видно что у нас много пропущенных значений

```python
outlier = data[['year_of_release','audience_count','imdb_rating','tomatometer_rating','audience_rating','tomatometer_count']]
outlier
```

```python
# Определить нижние пределы для выбросов
Q1 = outlier.quantile(0.25)
Q1
```

```python
# Определить верхние пределы выбросов
Q3 = outlier.quantile(0.75)
Q3
```

```python
# IQR-оценки
IQR = Q3-Q1
# Отфильтрованные данные, из которых удалены выбросы
data_filtered = outlier[~((outlier < (Q1 - 1.5 * IQR)) |(outlier > (Q3 + 1.5 * IQR))).any(axis=1)]
data_filtered
```

```python
index_list = list(data_filtered.index.values)
```

```python
# Хранить данные без выбросов в отдельном DataFrame.
data_filtered = data[data.index.isin(index_list)]
data_filtered.head()
```

```python
# создать гистограмму с отфильтрованным выбросом
filtered_outlier = data_filtered[['year_of_release','audience_count','imdb_rating','tomatometer_rating','audience_rating','tomatometer_count']]

for col in filtered_outlier:
    data_filtered.hist(col, bins=100)
```

Сравним значения с выбросами и без и сделаем выводы

```python
# Сравнение годов с выбросами и без них (гистограмма)
print (data.hist('year_of_release', bins=100))
print (data_filtered.hist('year_of_release', bins=100))
```

1) Видно что снимали больше фильмов в старые года
2) Более раздробленный график, который дает больше информации

```python
# Сравнение годов с выбросами и без них (гистограмма)
print (data.hist('audience_count', bins=100))
print (data_filtered.hist('audience_count', bins=100))
```

1) Выброс был уменьшен с 35000000 до 175000. Гистограммы просты для понимания.
2) Заметим что у большей части фильмов была не такая уж и большая оценочная аудитория.

Проанализируем количество фильмов разных возрастных рейтингов и средний рейтинг для каждого возрастного рейтинга. Выясним какое количество фильмов выпускают студии с разными возрастными рейтингами.

```python
# Создание переменной с возрастным рейтингом и средним imdb рейтингом
avg_imdb = data_filtered.pivot_table(index='content_rating', values='imdb_rating', aggfunc='mean')
avg_imdb.sort_values(ascending=False ,by='imdb_rating')
```

```python
# Создание переменной с типом и количеством объявлений
ads = data_filtered.pivot_table(index='content_rating', values='film_studio_or_producers', aggfunc='count')
ads = ads.sort_values(ascending=False, by ='film_studio_or_producers' )
ads = ads.reset_index()
ads
```

```python
ads.plot(grid=True, figsize=(20, 7), kind = 'bar', x='content_rating',y='film_studio_or_producers');
```

Средний рейтинг для каждого типа показывает, что с самым высоким рейтингом получаются фильмы NR и PG, скорее всего ээто связано с тем, что такие фильмы являются более просматривемыми, NR - фильмы которые не оценены рейтингом и таких много, а PG - фильмы которые можно смотреть со взрослыми без ограничения по возрасту. 

Проверим рейтинговые факторы:
1) Год

4) Время просмотра

Отфильтруем два популярных возрастных рейтинга

```python
NR = data_filtered.query('content_rating == "NR"')
PG = data_filtered.query('content_rating == "PG"')
```

```python
sns.lmplot(x="year_of_release", y="imdb_rating", data=NR)
plt.title('NR')
sns.lmplot(x="year_of_release", y="imdb_rating", data=PG)
plt.title('PG')
```

```python
age_NR = NR[['year_of_release','imdb_rating']]
corr_NR = age_NR.corr()
age_PG = PG[['year_of_release','imdb_rating']]
corr_PG = age_PG.corr()
```

```python
round(corr_NR,3)
```

```python
round(corr_PG,3)
```

```python
sns.lmplot(x="audience_count", y="audience_rating", data=NR, height = 5, aspect =3)
plt.title('NR')
sns.lmplot(x="audience_count", y="audience_rating", data=PG, height = 5, aspect =3)
plt.title('PG')
```

Можно заметить что рейтинг сильно зависит от числа опрашиваемых людей
